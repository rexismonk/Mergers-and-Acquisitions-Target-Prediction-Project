```python
import pandas as pd
import numpy as np


pd.set_option('display.max_columns', 100)
pd.set_option('display.width', 1000)
```

```python
# Define file paths 
investment_path = 'investments_VC.csv'
acquisition_path = 'Acquisitions.csv'

# Load the datasets

try:
    df_investments = pd.read_csv(investment_path, encoding='latin1')
    df_acquisitions = pd.read_csv(acquisition_path)
    print("Files loaded successfully!")
except FileNotFoundError as e:
    print(f"Error: {e}")
    print("Please ensure the CSV files are in a 'data/raw' folder relative to the 'notebooks' directory.")
```
